```python
# Replacing categorical values for 'Crop', 'Region', 'Soil_Type', etc.
data.replace({'Crop': {'Maize': 0, 'Rice': 1, 'Barley': 2, 'Wheat': 3, 'Cotton': 4, 'Soybean': 5}}, inplace=True)

data.replace({'Region': {'North': 0, 'West': 1, 'South': 2, 'East': 3}}, inplace=True)

data.replace({'Soil_Type': {'Sandy': 0, 'Loam': 1, 'Chalky': 2, 'Silt': 3, 'Clay': 4, 'Peaty': 5}}, inplace=True)

data.replace({'Weather_Condition': {'Sunny': 0, 'Rainy': 1, 'Cloudy': 2}}, inplace=True)

# Replace booleans (True/False) with integers (1/0)
data.replace({'Fertilizer_Used': {False: 0, True: 1}}, inplace=True)

data.replace({'Irrigation_Used': {False: 0, True: 1}}, inplace=True)
```

```python
print(data.head())
```

```python
x=data.drop(columns='Yield_tons_per_hectare',axis=1)
x.head()
```

```python
y=data['Yield_tons_per_hectare']
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=2,test_size=0.2)
```

```python
print(x.shape,xtrain.shape,xtest.shape)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import r2_score
from sklearn import metrics
```

```python
regressor = LinearRegression()
regressor.fit(xtrain,ytrain)
```

```python
training_data_prediction = regressor.predict(xtrain)
```

```python
r2_train = metrics.r2_score(ytrain, training_data_prediction)
print(r2_train)
```

```python
test_data_prediction = regressor.predict(xtest)
```

```python
r2_test = metrics.r2_score(ytest, test_data_prediction)
print(r2_train)
```

```python
ypred=regressor.predict(xtest)
ypred
```

```python
score=r2_score(ytest,ypred)
score
```

KNN

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
knn.fit(xtrain,ytrain)
```

```python
ypred=knn.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(ytest,ypred)
score
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
DT=DecisionTreeRegressor()
DT.fit(xtrain,ytrain)
```

```python
ypred=DT.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(ytest,ypred)
score
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rfc = RandomForestRegressor(n_estimators=20, criterion='squared_error', max_depth=5, max_features=5)
# help(rfc)
rfc.fit(xtrain,ytrain)
```

```python
ypred=rfc.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score
```

```python
print('Accuracy score :',r2_score(ytest,ypred))
```
